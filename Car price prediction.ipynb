# Train a Model

```python
from sklearn.model_selection import train_test_split
```

```python
cars.columns
```

```python
X = cars[['Present_Price', 'Driven_kms']]
y = cars[['Selling_Price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
model = LinearRegression()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model.fit(X_train_scaled,y_train)
```

```python
predict_lm = model.predict(X_test_scaled)
```

```python
sns.set_palette('bone')
plt.scatter(y_test,predict_lm)
```

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test,predict_lm)
```

```python
MSE = metrics.mean_squared_error(y_test,predict_lm)
```

```python
RMSE = np.sqrt(metrics.mean_squared_error(y_test,predict_lm))
```

```python
print("MAE:",MAE)
print("MSE:",MSE)
print("RMSE:",RMSE)
```

```python
sns.displot(y_test-predict_lm)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=1.0)
```

```python
lasso_model.fit(X_train_scaled, y_train)
```

```python
predict_lasso = lasso_model.predict(X_test_scaled)
```

```python
MAE = metrics.mean_absolute_error(y_test,predict_lasso)
```

```python
MSE = metrics.mean_squared_error(y_test,predict_lasso)
```

```python
RMSE = np.sqrt(metrics.mean_squared_error(y_test,predict_lasso))
```

```python
print("MAE:",MAE)
print("MSE:",MSE)
print("RMSE:",RMSE)
```

```python
sns.set_palette('cool')
plt.scatter(y_test,predict_lasso)
```

## Gradiant Boosting 

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gradient_boosting_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=101)
```

```python
gradient_boosting_model.fit(X_train_scaled, y_train)
```

```python
predict_gb = gradient_boosting_model.predict(X_test_scaled)
```

```python
MAE = metrics.mean_absolute_error(y_test,predict_gb)
MSE = metrics.mean_squared_error(y_test,predict_gb)
RMSE = np.sqrt(metrics.mean_squared_error(y_test,predict_gb))

print("MAE:",MAE)
print("MSE:",MSE)
print("RMSE:",RMSE)
```

```python
sns.set_palette('copper_r')
plt.scatter(y_test,predict_gb)
```
